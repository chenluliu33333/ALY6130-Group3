# Qualitative Risk Analysis

Summarises the 10 approved risks and their ERM categories.

```python
import pandas as pd

risks = [
    'Competitor AI Advantage',
    'AI Warehouse Strategy Failure',
    'Cybersecurity & Data Breach',
    'AI Forecasting Failure',
    'System Integration Failure',
    'Cloud Service Outage',
    'Workforce Adoption Failure',
    'Cost Overrun & Delayed ROI',
    'Regulatory & Privacy Non-Compliance',
    'AI Governance & Ethics Failure'
]

qualitative = pd.DataFrame({
    'Rank': range(1, 11),
    'Risk': risks,
    'ERM Category': ['Strategic','Strategic','Information Security','Technology','Technology','Technology','People','Financial','Compliance','Governance']
})
display(qualitative)
```
