# Quantitative Risk Analysis

Calculates expected loss ranking and total portfolio exposure from the processed risk summary.

```python
import pandas as pd
from pathlib import Path

quant = pd.read_csv(Path('../data/processed/quantitative_risk_summary.csv'))
quant['Expected Loss Rank'] = quant['expected_loss_million_usd'].rank(ascending=False, method='dense').astype(int)
display(quant.sort_values('expected_loss_million_usd', ascending=False))
print('Portfolio expected loss:', round(quant['expected_loss_million_usd'].sum(), 2), 'million USD')
```
